```python
(
    gapminder
    .filter(['continent', 'country', 'lifeExp', 'year'])
    .pivot(columns = 'year',
           values = 'lifeExp',
           index = ['continent', 'country'])
    .reset_index()
    .melt(id_vars = ['country', 'continent'],
         value_name = 'lifeExp')
)
```
